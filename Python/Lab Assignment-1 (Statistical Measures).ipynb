## Q1:

### You have the results of various bee species. The number of bees observed was as follows:	
### Buff Tail: 10 1 37 5 12
### Garden Bee: 8 3 19 6 4
### Red Tail: 18 9 1 2 4
### Honey Bee: 12 13 16 9 10
### Carder Bee: 8 27 6 32 23
### **a)**	Make simple numeric vector. Join the bee vectors together to make a data frame. Each row of the resulting data frame relates to a specific plant so you could assign names to the rows. The plant names are Thistle, Vipers Bugloss, Golden Rain, Yellow Alfalfa, and Blackberry.

```python
# Importing libraries 
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# Creating vectors corresponding to various bee species
buff_tail = [10,1,37,5,12]
garden_bee = [8,3,19,6,4]
red_tail = [18,9,1,2,4]
honey_bee = [12,13,16,9,10]
carder_bee = [8,27,6,32,23]
```

```python
# Creating vectors corresponding to plant names
plant_names = ['Thistle','Vipers Bugloss','Golden Rain','Yellow Alfalfa','Blackberry']
```

<br>

### **b)**	Make the matrix object from the original data.

```python
# Converting the vectors formed into dataframe
bee_df = pd.DataFrame({"Plant Name": plant_names,
                         "Buff Tail": buff_tail,
                         "Garden Bee": garden_bee,
                         "Red Tail": red_tail,
                         "Honey Bee": honey_bee,
                         "Carder Bee": carder_bee})

print(bee_df)
```

### **c)**	Display the data for the Blackberry only.

```python
# Printing the tuple for which "Plant Name" == "Blackberry" 
print(bee_df[bee_df["Plant Name"] == "Blackberry"])
```

### **d)**	Display the data for Golden Rain and Yellow Alfalfa.

```python
# Printing the tuple for which "Plant Name" == "Golden Rain" OR "Plant Name" == "Yellow Alfalfa" 
print(bee_df[(bee_df["Plant Name"] == "Golden Rain") | (bee_df["Plant Name"] == "Yellow Alfalfa")])
```

<br> <br> <br> <br> 

### **e)**	Display the data for the Red Tail bee only.

```python
# Printing two columns, namely "Plant Name" and "Red Tail"
print(bee_df[["Plant Name","Red Tail"]])
```

### **f)**	Reorder the data so that the Buff Tail column is sorted in decreasing abundance.

```python
# Sorting the dataset on the basis of "Buff Tail" column in descending order  
print(bee_df.sort_values(by=["Buff Tail"],ascending=False))
```
