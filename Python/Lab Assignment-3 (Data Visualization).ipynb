# WAP to demonstrate graphical display of data distribution using: bar plot, box plot, and area plot.

## Bar Plot                      
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.
A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value.
Matplotlib API provides the bar() function that can be used in the MATLAB style use as well as object oriented API.

```python
from matplotlib import pyplot as plt 
```

```python
number = range(1,11)
girth = df['Girth'].head(10) 
  
# Figure Size 
fig = plt.figure(figsize =(9, 6)) 
  
# Horizontal Bar Plot 
plt.bar(number[0:10], girth[0:10]) 

# Labelling of the Plot
plt.ylabel("Girth of tree (in inches)") 
plt.xlabel("Tree number") 
plt.title("Bar plot of Girth of first 10 trees of dataset") 

# Show value of Girth along each bar in the plot
for index, value in enumerate(girth):
    plt.text(index+1, value, str(value))

# Show grid lines along y axis
plt.grid(axis='y', alpha=0.75)

# Show Plot 
plt.show() 
```

## Box Plot                             
A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.   
A box plot consist of 5 things: Minimum, First Quartile or 25%, Median (Second Quartile) or 50%, Third Quartile or 75% and Maximum; collectively called as **Five-number summary**.

```python
import matplotlib.pyplot as plt 
```

```python
# Figure Size 
fig = plt.figure(figsize =(9, 6))

# Creating plot
boxdata = [df['Girth'], df['Height'], df['Volume']]
plt.boxplot(boxdata, labels = ['Girth', 'Height', 'Volume']) 

# Show grid lines along y axis
plt.grid(alpha=0.75)

# Labelling of the Plot
plt.ylabel("Value in respective units") 
plt.xlabel("Attributes") 
plt.title("Boxplot of the dataset")

# Show Plot 
plt.show()
```

An **outlier** is a data point that differs significantly from other observations. It may be due to variability in the measurement or it may indicate experimental error; the latter are sometimes excluded from the data set. It can be seen that the attributes Ozone and Wind have outliers. Outliers are the data points above and below the upper fence and lower fence of the box plot respectively.                   
**Removing outliers** or handling outliers is necessary as they can cause serious problems in statistical analyses.     
Inter-quartile range = Third Quartile value - First Quartile value        
Value of upper fence = Third Quartile Value + 1.5 \* Inter-quartile range       
Value of lower fence = First Quartile Value - 1.5 \* Inter-quartile range        
To remove outliers, all data poits above the upper fence and below the lower fence need to be removed.

## Area Plot                 

An Area Plot is an extension of a Line Chart. Area Plot is obtained by filling the region between the Line Chart and the axes with a color. When more than one Area Plot is shown in the same graph, each area plot is filled with a different color.       
Due to the color-fill effect of an area plot, the quantum and the trend of the variable is distinctly visible without making much effort.

```python
# Figure Size 
fig = plt.figure(figsize =(9, 6))

# Defining the variables for axes
x=range(1,32)
y=[ df['Girth'], df['Height'], df['Volume']]
 
# Basic stacked area chart.
plt.stackplot(x,y, labels=['Girth','Height', 'Volume'])
plt.legend(loc='upper left')

# Labelling of the Plot
plt.ylabel("Girth(inches), Height(ft), Volume(cubic ft)") 
plt.xlabel("Tree number") 
plt.title("Stacked Area Plot of trees dataset")

# Show Plot
plt.show()
 
# This is just a representational graph, the units of Girth, Height and Volume are inches, feets and cubic ft respectively.
```

## Quantile-Quantile Plot (Q-Q Plot)                  

When the quantiles of two variables are plotted against each other, then the plot obtained is known as quantile – quantile plot or qqplot. This plot provides a summary of whether the distributions of two variables are similar or not with respect to the locations.     
- If all point of quantiles lie on or close to straight line at an angle of 45 degree from x – axis. It indicates that two samples have similar distributions. And in practice it is always not possible to get such a 100 percent clear straight line.                     
- If the y – quantiles are lower than the x – quantiles. It indicates y values have a tendency to be lower than x values. And in practice it is not always possible to get 100 percent as shown above but the plot looks as shown below.    
- If the x – quantiles are lower than the y – quantiles. It indicates x values have a tendency to be lower than the y values.                 
- There can be a breakpoint up to which the y – quantiles are lower than the x – quantiles and after that point the y – quantiles are higher than the x – quantiles.

```python
# It compares the quantiles of our data with the quantiles of the desired distribution (default to normal distrbution).
import statsmodels.api as sm
import pylab as py

# To compare 50 random floating numbers between 0 and 1 with x=y line
data_points = np.random.normal(0,1,50)

# Q-Q plot
sm.qqplot(data_points, line ='45')
py.show()

"""
line = {None, “45”, “s”, “r”, “q”}
Options for the reference line to which the data is compared:
“45” - 45-degree line
“s” - standardized line, the expected order statistics are scaled by the standard deviation of the given sample,
      and have the mean added to them
“r” - A regression line is fit
“q” - A line is fit through the quartiles.
None - by default no reference line is added to the plot.
"""
```

```python
# Q-Q plot comparing the Height varible of trees dataset with its own regression line.
sm.qqplot(df['Height'], line ='r')
py.show()
```

## Histogram                           
A histogram is basically used to represent data provided in a form of some groups.It is accurate method for the graphical representation of numerical data distribution.It is a type of bar plot where X-axis represents the bin ranges while Y-axis gives information about frequency.

```python
# Figure Size 
fig = plt.figure(figsize =(9, 6))

# Plotting histogram for Height of trees
plt.hist(df['Height'],color = "red")

# Labelling of the Plot
plt.ylabel("Frequency") 
plt.xlabel("Height of trees (in feet)") 
plt.title("Histogram of Height of Trees") 

# Show Plot
plt.show() 
```

## Scatter Plot                      
A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

```python
# Figure Size 
fig = plt.figure(figsize =(9, 6))

# Plotting scater plot with Height of trees on y-axis and girth of trees on x-axis
plt.scatter(df['Girth'], df['Height'])

# Labelling of the Plot
plt.ylabel("Height of trees (in feet)") 
plt.xlabel("Girth of trees (in inches)") 
plt.title("Scatter plot of Girth v/s Height of Trees") 

# Show Plot
plt.show()
```
